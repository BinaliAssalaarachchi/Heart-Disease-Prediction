```python
# feature selection using Chi-Square and ANOVA F-value
```

```python
# Chi-Square for categorical features

# Encode target for classification
label_enc = LabelEncoder()
y_class = label_enc.fit_transform(df['Heart Disease Status'])

# Features (remove target and numerical features)
X_class = df.drop(['Age', 'Blood Pressure', 'Cholesterol Level', 'BMI', 'Sleep Hours', 'Triglyceride Level','Fasting Blood Sugar', 'CRP Level', 'Homocysteine Level', 'Heart Disease Status'], axis=1)

# Fit Chi-Square for ALL features
chi_selector = SelectKBest(score_func=chi2, k='all')
chi_selector.fit(X_class, y_class)

# Build a table of per-feature Chi2 stats
chi_table = pd.DataFrame({
    'Feature': X_class.columns,
    'Chi2_Score': chi_selector.scores_,
    'p_value': chi_selector.pvalues_
}).sort_values(by='Chi2_Score', ascending=False)

print("Top 5 features by Chi-Square:")
print(chi_table.head(5))

print("\nFull Chi-Square table:")
print(chi_table)
```

```python
# ANOVA F-value for numerical features

# keep only numerical features
X= df[numeric_cols]

# apply ANOVA F-value
anova_selector = SelectKBest(score_func=f_classif, k='all')
anova_selector.fit(X, y_class)

# create DataFrame with scores and p-values
anova_scores_df = pd.DataFrame({
    'Feature': X.columns,
    'F_Score': anova_selector.scores_,
    'p_value': anova_selector.pvalues_
}).sort_values(by='F_Score', ascending=False, ignore_index=True)

# print full table
print("\nFull ANOVA scores and p-values:")
print(anova_scores_df)

# print top 5 features
top_features = anova_scores_df.head(5)
print("\nTop 5 features (ANOVA F-value):")
print(top_features)
```

```python
# feature selection using Variance Threshold and PCA dimensionality reduction
```

```python
# feature selection (Variance Threshold)
X=df.drop('Heart Disease Status', axis=1)   # Drop target column
selector = VarianceThreshold(threshold=0.005)  # remove near-constant features
X_sel = selector.fit_transform(X)

selected_features = X.columns[selector.get_support()].tolist()
X_sel_df = pd.DataFrame(X_sel, columns=selected_features)

print("\nSelected features after Variance Threshold:")
print(selected_features)
```

```python
# all features are selected, which means they have a significant variance in the dataset
```

```python
# PCA (manual eigen decomposition)

# Correlation matrix
corr_matrix = np.corrcoef(X_sel.T)

# Eigen decomposition
eigenvalues, eigenvectors = np.linalg.eigh(corr_matrix)

# sort eigenvalues/eigenvectors
sorted_idx = np.argsort(eigenvalues)[::-1]
eigenvalues_sorted = eigenvalues[sorted_idx]
eigenvectors_sorted = eigenvectors[:, sorted_idx]

# explained variance %
explained_variance_pct = (eigenvalues_sorted / np.sum(eigenvalues_sorted)) * 100
cum_explained_variance_pct = np.cumsum(explained_variance_pct)

# projection (scores)
k = 2
X_reduced = X_sel @ eigenvectors_sorted[:, :k]
```

```python
# display the summary of the PCA results

summary = pd.DataFrame({
    "PC": [f"PC{i+1}" for i in range(len(eigenvalues_sorted))],
    "Eigenvalue": eigenvalues_sorted,
    "Explained Variance (%)": explained_variance_pct.round(2),
    "Cumulative (%)": cum_explained_variance_pct.round(2)
})
print("\nPCA Results:")
print(summary)
```

```python
# this shows that all features explain the same percentage of variance in the data (approximately 5%) 
# no single PC dominates
```

```python
# display Scree + Cumulative plot

plt.figure(figsize=(8,5))
plt.plot(range(1, len(eigenvalues_sorted)+1), explained_variance_pct, marker='o', label="Individual")
plt.plot(range(1, len(eigenvalues_sorted)+1), cum_explained_variance_pct, marker='s', label="Cumulative")
plt.title("Explained Variance (%) by Principal Components")
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance (%)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# display 2D Scatter Plot (PC1 vs PC2)

plt.figure(figsize=(6,5))
plt.scatter(X_reduced[:,0], X_reduced[:,1])
plt.title("PCA Projection (PC1 vs PC2)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()
```
