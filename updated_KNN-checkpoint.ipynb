```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import StratifiedKFold
from imblearn.over_sampling import SMOTE
```

```python
# Load the pre-processed dataset
data = pd.read_csv("pre-processed_dataset.csv")
```

```python
# Separate features (X) and target (y)
X = data.drop("Heart Disease Status", axis=1)
y = data["Heart Disease Status"]
```

```python
# Oversampling using SMOTE
smote = SMOTE(random_state=42)
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

accuracy_scores = []
f1_scores = []
precision_scores = []
recall_scores = []
auc_scores = []
confusion_matrices = []

for fold, (train_index, test_index) in enumerate(skf.split(X, y)):
    print(f"Fold {fold+1}")
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Apply SMOTE to the training data
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

    # Train the initial (untuned) KNN model with n_neighbors=5
    model_knn_initial = KNeighborsClassifier(n_neighbors=5)
    model_knn_initial.fit(X_train, y_train)

    # Make predictions
    y_pred = model_knn_initial.predict(X_test)
    y_proba = model_knn_initial.predict_proba(X_test)[:, 1]

    # Calculate metrics
    accuracy_scores.append(accuracy_score(y_test, y_pred))
    f1_scores.append(f1_score(y_test, y_pred))
    precision_scores.append(precision_score(y_test, y_pred))
    recall_scores.append(recall_score(y_test, y_pred))
    auc_scores.append(roc_auc_score(y_test, y_proba))
    confusion_matrices.append(confusion_matrix(y_test, y_pred))
```

```python
# Evaluate the initial model
print("Initial KNN Model Metrics:")
print(f"Average Accuracy: {np.mean(accuracy_scores):.4f}")
print(f"Average F1 Score: {np.mean(f1_scores):.4f}")
print(f"Average Precision: {np.mean(precision_scores):.4f}")
print(f"Average Recall: {np.mean(recall_scores):.4f}")
print(f"Average AUC: {np.mean(auc_scores):.4f}")
```

```python
# Display confusion matrix for last fold
disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrices[-1], display_labels=["No Heart Disease", "Heart Disease"])
disp.plot(cmap='Blues')
plt.title("Confusion Matrix for Last Fold - Initial KNN Model")
plt.show()
```

```python
# Define parameter grid for GridSearchCV and RandomizedSearchCV
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}
```

```python
# Initialize KNN classifier
knn = KNeighborsClassifier()
```

```python
# Perform GridSearchCV
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
```

```python
# Get best models
best_grid_model = grid_search.best_estimator_
```

```python
# Evaluate best GridSearchCV model
y_pred_grid = best_grid_model.predict(X_test)
print("GridSearchCV Best Parameters:", grid_search.best_params_)
print("GridSearchCV Model Metrics:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_grid)}")
print(f"Precision: {precision_score(y_test, y_pred_grid)}")
print(f"Recall: {recall_score(y_test, y_pred_grid)}")
print(f"F1 Score: {f1_score(y_test, y_pred_grid)}")
print(f"ROC AUC: {roc_auc_score(y_test, best_grid_model.predict_proba(X_test)[:, 1])}")
print("-" * 50)
```

```python
# Assuming best_grid_model is defined from a previous GridSearchCV
y_pred_grid = best_grid_model.predict(X_test)
cm_grid = confusion_matrix(y_test, y_pred_grid)
disp_grid = ConfusionMatrixDisplay(confusion_matrix=cm_grid, display_labels=["No Heart Disease", "Heart Disease"])
disp_grid.plot(cmap='Blues')
plt.title("Confusion Matrix - GridSearchCV Tuned KNN Model")
plt.show()

# Calculate metrics for the tuned GridSearchCV model on the test set
y_pred_grid = best_grid_model.predict(X_test)
y_proba_grid = best_grid_model.predict_proba(X_test)[:, 1]

accuracy_grid = accuracy_score(y_test, y_pred_grid)
f1_grid = f1_score(y_test, y_pred_grid)
precision_grid = precision_score(y_test, y_pred_grid)
recall_grid = recall_score(y_test, y_pred_grid)
auc_grid = roc_auc_score(y_test, y_proba_grid)
```

```python
# Calculate accuracy for the tuned GridSearchCV model on the test set
y_pred_grid = best_grid_model.predict(X_test)
accuracy_grid = accuracy_score(y_test, y_pred_grid)

# Prepare data for bar graph (only accuracy)
metrics = ['Accuracy']
base_means = [np.mean(accuracy_scores)]
tuned_means = [accuracy_grid]

# Create bar graph
x = np.arange(len(metrics))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(6, 6))
rects1 = ax.bar(x - width/2, base_means, width, label='Base Model (CV Average)', color='#1f77b4')
rects2 = ax.bar(x + width/2, tuned_means, width, label='Tuned GridSearch Model (Test Set)', color='#ff7f0e')

# Add labels, title, and custom x-axis tick labels
ax.set_ylabel('Accuracy Score')
ax.set_title('Accuracy Comparison of Base and Tuned KNN Models')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.set_ylim(0, 1)  # Assuming accuracy scores are between 0 and 1
ax.legend()

# Function to add value labels on bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.4f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
```
