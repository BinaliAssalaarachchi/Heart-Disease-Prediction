```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv('pre-processed_dataset.csv')
```

```python
#Separate Features and Target
X = df.drop('Heart Disease Status', axis=1)
y = df['Heart Disease Status']
```

```python
#Apply SMOTE Oversampling
import os
os.environ["LOKY_MAX_CPU_COUNT"] = "4" 

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

X_resampled = np.array(X_resampled)
y_resampled = np.array(y_resampled)
```

```python
#Stratified K-Fold Setup
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
#Define Model Variants
model_configs = [
    {'name': 'Model 1: Default RF', 'params': {}},
    {'name': 'Model 2: Tuned RF', 'params': {'n_estimators': 200, 'max_depth': 10}},
    {'name': 'Model 3: Aggressive RF', 'params': {'n_estimators': 300, 'max_depth': 20, 'min_samples_split': 5}}
]
```

```python
#Train and Evaluate Each Model
for config in model_configs:
    print(f"\n🔍 Evaluating {config['name']}")
    acc, prec, rec, f1, auc = [], [], [], [], []
    last_conf_matrix = None

    for train_idx, test_idx in skf.split(X_resampled, y_resampled):
        X_train, X_test = X_resampled[train_idx], X_resampled[test_idx]
        y_train, y_test = y_resampled[train_idx], y_resampled[test_idx]

        model = RandomForestClassifier(class_weight='balanced', random_state=42, **config['params'])
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_proba = model.predict_proba(X_test)[:, 1]

        acc.append(accuracy_score(y_test, y_pred))
        prec.append(precision_score(y_test, y_pred, zero_division=0))
        rec.append(recall_score(y_test, y_pred, zero_division=0))
        f1.append(f1_score(y_test, y_pred, zero_division=0))
        auc.append(roc_auc_score(y_test, y_proba))
        last_conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
#Print Metrics
print(f"Accuracy     : {np.mean(acc):.4f}")
print(f"Precision    : {np.mean(prec):.4f}")
print(f"Recall       : {np.mean(rec):.4f}")
print(f"F1 Score     : {np.mean(f1):.4f}")
print(f"ROC AUC      : {np.mean(auc):.4f}")
print("Confusion Matrix (Last Fold):")
print(last_conf_matrix)
```
